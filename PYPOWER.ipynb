```python
from pypower.api import case14, runpffrom pypower.idx_brch import F_BUS, T_BUSfrom pypower.idx_bus import BUS_I
```

```python
import networkx as nx
```

```python
ppc = case14()
```

```python
g = nx.Graph()i = ppc['bus'][:, BUS_I].astype(int)g.add_nodes_from(i, bgcolor='green')fr = ppc['branch'][:, F_BUS].astype(int)to = ppc['branch'][:, T_BUS].astype(int)g.add_edges_from(zip(fr, to), color='blue')nx.draw(g)
```

```python
ppc = runpf(ppc)
```
